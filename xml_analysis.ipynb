### merging the xml files
Thanks ChatGPT for the code: https://chat.openai.com/share/fb3bab7c-a44c-48e0-9fec-3c6ac95c0355

```python
import os
import xml.etree.ElementTree as ET

def merge_xml_files(directory):
    xml_files = [os.path.join(directory, file) for file in os.listdir(directory) if file.endswith(".xml")]
    xml_files.sort()  # ensure files are processed in order

    first_file = xml_files.pop(0)

    # parse the first file and get its root to which other files' roots' children will be appended
    master_tree = ET.parse(first_file)
    master_root = master_tree.getroot()

    for xml_file in xml_files:
        tree = ET.parse(xml_file)
        root = tree.getroot()

        for child in root:
            master_root.append(child)

    # write the final tree to an xml file
    master_tree.write("all_merged_v0.xml")


# Call the function with the directory containing the XML files
merge_xml_files("/Users/kuboi/fandom-chatbot/ScrapeFandom-main/filter_test_0")
```
